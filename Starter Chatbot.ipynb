```python
import nltk
from nltk.chat.util import Chat, reflections
import random
nltk.download('punkt')
```

```python
intents = {
    'greeting': ['Hello!', 'Hi there!', 'Greetings!', 'How can I help you?'],
    'goodbye': ['Bye! Take care!', 'See you later!', 'Goodbye!'],
    'thanks': ['You\'re welcome!', 'Glad to help!', 'Anytime!'],
    'name': ['I am a chatbot created for demonstration purposes.'],
    'age': ['I am ageless!'],
    'unknown': ['Sorry, I do not understand that.']
}
```

```python
def get_response(user_input):
    # streamine the  user input
    user_input = user_input.lower()

    # word recongnition
    if any(word in user_input for word in ['hi', 'hello', 'hey']):
        return random.choice(intents['greeting'])
    elif any(word in user_input for word in ['bye', 'quit', 'exit']):
        return random.choice(intents['goodbye'])
    elif 'thank' in user_input:
        return random.choice(intents['thanks'])
    elif 'name' in user_input:
        return random.choice(intents['name'])
    elif 'age' in user_input:
        return random.choice(intents['age'])
    else:
        return random.choice(intents['unknown'])
```

```python
print("Chat with me! Type 'quit' to exit.")
while True:
    user_input = input("You: ")
    if user_input.lower() in ['quit', 'exit']:
        print("Chatbot:", random.choice(intents['goodbye']))
        break
    response = get_response(user_input)
    print(f"Chatbot: {response}")
```
